```python
import sys
sys.path.append('..')

import pandas as pd
import numpy as np
import yaml
from sklearn.metrics import roc_auc_score

from src.data import load_train, preprocess
from src.features import get_features
from src.models import cross_validate
```

```python
cfg = yaml.safe_load(open('../configs/baseline.yaml'))
train = load_train()
X, y = preprocess(train, cfg['data']['target_col'], cfg['data']['drop_cols'])
X = get_features(X)
```

```python
models, oof_preds, fold_scores = cross_validate(X, y, cfg['hyperparams'], n_splits=cfg['cv']['n_splits'])
print(f'Fold AUCs: {[f"{s:.4f}" for s in fold_scores]}')
print(f'OOF AUC:  {roc_auc_score(y, oof_preds):.5f}')
```
